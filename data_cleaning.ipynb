```python
import pandas as pd
```

```python
data = pd.ExcelFile ("Informe Resumen Enero 21.xlsx")
```

```python
product_historic = pd.read_excel (data, "HISTORICO X PRODUCTO")
```

```python
product_historic.head()
```

```python
def separate_data(dataset):
    
    data_divided = []
    num_columns1 = [1,2,3,4,5]
    index = dataset[dataset.columns [[0]]]
    set_1 = dataset[dataset.columns [[num_columns1]]]
    set_2 = dataset[dataset.columns [[i+5 for i in num_columns1 ]]]
    set_3 = dataset[dataset.columns [[i+10 for i in num_columns1 ]]]
    set_4 = dataset[dataset.columns [[i+15 for i in num_columns1 ]]]
    set_5 = dataset[dataset.columns [[i+20 for i in num_columns1 ]]]
    set_6 = dataset[dataset.columns [[26,27]]]
    set_7 = dataset[dataset.columns [[28,29]]]
    set_8 = dataset[dataset.columns [[30,31]]]
    total = dataset[dataset.columns [[32]]]
    data_divided.append (index)
    data_divided.append (set_1)
    data_divided.append (set_2)
    data_divided.append (set_3)
    data_divided.append (set_4)
    data_divided.append (set_5)
    data_divided.append (set_6)
    data_divided.append (set_7)
    data_divided.append (set_8)
    data_divided.append (total)
    return (data_divided)
    
   
    
    
```

```python
data_divided = separate_data(product_historic)
```

```python
data_divided[0].columns = data_divided[0].iloc[0]
data_divided[0] = data_divided[0][1:]
```

```python
data_cleaned = []
for d in data_divided[1:]:
    d.columns = d.loc[0]
    d = d[1:]
    x = pd.concat ([data_divided[0],d], axis = 1, sort=False)
    data_cleaned.append (x)
```

```python
data_ready = []
for d in data_cleaned:
    x =d.fillna(0)
    x = x.iloc[:,1:].astype(int)
    x.insert(0,"Mes", data_divided[0])
    data_ready.append (x)
    
```

```python
particulares_livianos = data_ready[0]
taxis = data_ready[1]
publicos_livianos = data_ready[2]
motos = data_ready[3]
preventivas_livianos = data_ready[4]
particulares_pesados = data_ready[5]
publicos_pesados = data_ready[6]
preventiva_pesados = data_ready[7]
```

```python
def rearrange_df (df):
    df [["Mes","Año"]] = df["Mes"].str.split("(", expand =True)
    df["Año"] = df["Año"].str.replace(")", "")
    cols = ["Año", "Mes", 80, 170, 13, "CELTA", "TOTAL"]
    df = df [cols]
    df.rename (columns ={"Año":"anio", "Mes":"mes", 80: "cda_80", 170: "cda_170", 13:"cda_13", "CELTA": "cda_celta", "TOTAL":"total"}, inplace = True) 
    df = df[:-4]
    return df

def rearrange_df_pesados (df):
    df [["Mes","Año"]] = df["Mes"].str.split("(", expand =True)
    df["Año"] = df["Año"].str.replace(")", "")
    cols = ["Año", "Mes", "CELTA", "TOTAL"]
    df = df [cols]
    df.rename (columns ={"Año":"anio", "Mes":"mes", "CELTA": "cda_celta", "TOTAL":"total"},inplace = True) 
    df = df[:-4]
    return df


```

```python
particulares_livianos = rearrange_df (particulares_livianos)
taxis = rearrange_df (taxis)
publicos_livianos = rearrange_df (publicos_livianos)
motos = rearrange_df (motos)
preventivas_livianos = rearrange_df (preventivas_livianos)
```

```python
particulares_pesados = rearrange_df_pesados (particulares_pesados)
publicos_pesados = rearrange_df_pesados (publicos_pesados)
preventiva_pesados = rearrange_df_pesados (preventiva_pesados)
```

```python
particulares_livianos.to_csv("particulares_livianos.csv", index=False)
taxis.to_csv("taxis.csv", index=False)
publicos_livianos.to_csv("publicos_livisnos.csv", index=False)
motos.to_csv("motos.csv", index=False)
preventivas_livianos.to_csv("prevetivas_livianos.csv", index=False)
particulares_pesados.to_csv("particulares_pesados.csv", index=False)
publicos_pesados.to_csv("publicos_pesados.csv", index=False)
preventiva_pesados.to_csv("preventiva_pesados.csv", index=False)
```
