```python
class Record(Base):
    __tablename__ = 'particulares_livianos'
    anio = Column(Integer, primary_key=True)
    mes = Column(String(255))
    cda_80 = Column(Integer)
    cda_170 = Column(Integer)
    cda_13 = Column(Integer)
    cda_celta = Column(Integer)
    total = Column(Integer)
```
